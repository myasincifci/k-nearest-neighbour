```python
#!pip install numpy
import numpy as np
```

```python
# BLUE
mean_b = [1,0]
cov_b  = [[1,0],[0,1]]

# ORANGE 
mean_o = [0,1]
cov_o  = [[1,0],[0,1]]
```

```python
#!pip install matplotlib
import matplotlib.pyplot as plt

#BLUE means
x_b, y_b = np.random.multivariate_normal(mean_b, cov_b, 10).T
#ORANGE means
x_o, y_o = np.random.multivariate_normal(mean_o, cov_o, 10).T
```

```python
# GENERATE OBSERVATIONS
import random 

# BLUE
r = random.randrange(10)
m_b = [x_b[r], y_b[r]]

x_ob, y_ob = np.random.multivariate_normal(m_b, [[1/5,0],[0,1/5]], 100).T

# ORANGE
r = random.randrange(10)
m_o = [x_o[r], y_o[r]]

x_oo, y_oo = np.random.multivariate_normal(m_o, [[1/5,0],[0,1/5]], 100).T
```

```python
# COMBINE
X_1 = list(x_ob) + list(x_oo)
X_2 = list(y_ob) + list(y_oo)
Y   = list(np.zeros(len(x_ob))) + list(np.ones(len(x_oo)))
```

```python
# PLOT
plt.scatter(X_1, X_2,c=Y)
plt.axis('equal')
plt.show()
print(len(X_1))
```
